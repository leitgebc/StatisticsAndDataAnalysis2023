```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('flowershapes.csv',delimiter=',').T
props = ['Sepal Length','Sepal Width','Petal Length','Petal Width']
```

```python
means,medians = np.mean(data,axis=1),np.median(data,axis=1)
stds = np.std(data,axis=1)

print('mean: ',means,'\nvariance: ',stds**2,'\nstandard deviation: ',stds,'\nmedian: ',medians)
print('covariance:\n',np.cov(data),'\ncorrelation:\n',np.corrcoef(data))
```

```python
fig, axs = plt.subplots(2, 2)
locs = [(0,0),(0,1),(1,0),(1,1)]
cols = ['tab:blue','tab:orange','tab:green','tab:red']
binnrs = [10,10,10,10]

for i in range(len(data)):
    counts, bins = np.histogram(data[i],bins=binnrs[i])
    axs[locs[i]].stairs(counts,bins,color=cols[i],fill=True)
    axs[locs[i]].set_title(props[i])
    axs[locs[i]].axvline(means[i],color='dimgrey',label='mean')
    axs[locs[i]].axvline(medians[i],color='dimgrey',ls='--',label='median')
    axs[locs[i]].errorbar(means[i],max(counts)/4,xerr=stds[i],color='dimgrey',capsize=4)
    axs[locs[i]].legend()
fig.tight_layout()
plt.show()
```
