```python
import numpy as np
import matplotlib.pyplot as plt
```

**Exercise 1**

Taylor expansion around $\mu_\phi=\langle\phi\rangle$: $$\sin\left(\frac{\phi}{2}\right)\approx\sin\left(\frac{\mu_\phi}{2}\right)+\frac{1}{2}\cos\left(\frac{\mu_\phi}{2}\right)(\phi-\mu_\phi)$$

$\implies$ error propagation for individual measurements $\phi_i$: $$V\left(\sin\left(\frac{\phi_i}{2}\right)\right)=\frac{1}{4}\cos^2\left(\frac{\phi_i}{2}\right)V(\phi_i)$$

```python
n = np.arange(1,10)
phi = 2*np.pi/360 * np.array([12,22,25,38,46,72,77,86,135]) #corresponds to averages
err_phi = 2*np.pi/360 * np.array([2,4,5,6,7,10,12,16,18]) #corresponds to std deviations
err_sin = abs(1/2 * np.cos(phi/2) * err_phi)

plt.errorbar(n, np.sin(phi/2), err_sin, marker = 'o', ls = '', capsize = 3)
plt.xlabel("$n$"); plt.ylabel(r"$\sin(\phi/2)$")
plt.grid(alpha = 0.5)
plt.show()
```

```python
d = 0.0711e-9/2 * n/np.sin(phi/2)
err_d = abs(0.0711e-9/2 * n/np.sin(phi/2)**2 * err_sin)

d_avg = sum(d/err_d**2) / sum(1/err_d**2) #weighted average
d_avg_err = np.sqrt(1/sum(1/err_d**2))
print(d_avg, "pm", d_avg_err)
```
