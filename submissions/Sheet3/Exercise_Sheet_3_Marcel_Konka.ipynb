1. a.)
Given that the mean for the Poisson distribution is $E(k)=\lambda$, we can calculate the variance V(k):

$V(k)= E(k^2)-E(k)^2 = \sum_{k=0}^\infty k^2\frac{e^{-\lambda}\lambda^k}{k!} - \lambda^2$

$= \sum_{k=1}^\infty k^2\frac{e^{-\lambda}\lambda^k}{k!} - \lambda^2 = \sum_{k=1}^\infty k\frac{e^{-\lambda}\lambda^k}{(k-1)!} - \lambda^2$

$= \lambda(\sum_{k=0}^\infty (k+1)\frac{e^{-\lambda}\lambda^k}{(k)!}) -\lambda^2 $

$= \lambda(\sum_{k=0}^\infty k\frac{e^{-\lambda}\lambda^k}{(k)!} + \sum_{k=0}^\infty \frac{e^{-\lambda}\lambda^k}{(k)!})  -\lambda^2 $

$= (\lambda\sum_{k=1}^\infty k\frac{e^{-\lambda}\lambda^k}{(k)!}+ \lambda e^{-\lambda}e^{\lambda} ) -\lambda^2$

$= (\lambda\sum_{k=1}^\infty \frac{e^{-\lambda}\lambda^k}{(k-1)!}+ \lambda)  -\lambda^2$

$= (\lambda\sum_{k=0}^\infty \frac{e^{-\lambda}\lambda^{k+1}}{k!}+ \lambda)  -\lambda^2$

$=( \lambda\sum_{k=0}^\infty \frac{e^{-\lambda}\lambda^{k+1}}{k!}+ \lambda)  -\lambda^2$

$= (\lambda^2 e^{-\lambda} e^{\lambda}+ \lambda)  -\lambda^2$

$= \lambda$

1.b.) Given a uniform distribution between $a-b$: with mean $E(x)= \frac{a+b}{2}$:

$V(x) = E(x^2)-E(x)^2 $

$=\int_a^b\frac{x^2}{b-a} - (\frac{a+b}{2})^2$

$=[\frac{x^3}{3(b-a)}]_a^b - (\frac{a+b}{2})^2$

$=\frac{4(b-a)(a^2+ab+b^2)}{12} \frac{3(a+b)^2}{12}$

$= \frac{4a^2+4ab+4b^2-3a^2-6ab-3b^2}{12}$

$=\frac{b^2-2ab+a^2}{12}$

$=\frac{(b-a)^2}{12}$



```python
#2a) Creating a plot of np.sin(x)^2-np.cos(x)^3
res = 10**6
x = np.linspace(0,np.pi*2,res)
y = np.ones(res)
for i in range(res):
    y[i] = np.sin(x[i])**2-np.cos(x[i])**3
plt.plot(x,y,color= "red")
plt.xlim(0,np.pi*2)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
#2.b-d)
x_rand = np.random.uniform(low=0, high=2*np.pi, size=(res,))
y_rand = np.random.uniform(low=-1, high=1, size=(res,))

x_accepted = np.zeros(res)
y_accepted = np.zeros(res)
Counter = 0

for i in range(res):
    y_test = np.sin(x_rand[i])**2-np.cos(x_rand[i])**3
    if y_rand[i] < y_test:
        x_accepted[Counter] = x_rand[i]
        y_accepted[Counter] = y_rand[i]
        Counter +=1
x_accepted = x_accepted[:Counter]
y_accepted = y_accepted[:Counter]
plt.scatter(x_accepted,y_accepted,linewidths = 0.001)
plt.plot(x,y,color = "red")
plt.legend(["Accepted Points","f(x)"])
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
#2.e)
bins = 200
plt.hist(x_accepted,bins = np.linspace(0,2*np.pi,bins), weights = np.ones(Counter)*bins/res)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
