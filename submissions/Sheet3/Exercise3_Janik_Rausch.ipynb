```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 200)
f = lambda x: np.sin(x)**2 - np.cos(x)**3 + 1
#I shifted f(x) by 1 to make it strictly positive

plt.plot(x, f(x))
plt.xlim(0, 2*np.pi)
plt.ylim(0)
plt.show()
```

```python
N = 1000000

rand_x = np.random.rand(N) * 2*np.pi
rand_y = np.random.rand(N) * 2

indices = np.where(rand_y < f(rand_x))
res = rand_x[indices]

plt.plot(x, f(x))
plt.scatter(res, rand_y[indices], s = 1, color = "tab:orange")
plt.xlim(0, 2*np.pi)
plt.ylim(0)
plt.show()
```

```python
counts, bins = np.histogram(res, bins = 60)
m = max(counts)

plt.plot(x, f(x) * m/2)
plt.stairs(counts, bins, fill = True)
plt.show()
```
