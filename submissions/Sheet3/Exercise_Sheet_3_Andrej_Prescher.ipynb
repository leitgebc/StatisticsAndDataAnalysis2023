```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 1000)
f = np.sin(x)**2-np.cos(x)**3

plt.figure(1)
plt.plot(x,f)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
np.random.seed(0)
randx = np.random.uniform(0, 2*np.pi, 1000000)
randy = np.random.uniform(-1, 1, 1000000)
```

```python
rand_var = []
rand_var_x = []
f_acc = []
for i in range(1000000):
    frac = (np.sin(randx[i])**2-np.cos(randx[i])**3)/1
    if randy[i] > frac:
        pass
    else:
        rand_var_x.append(randx[i])
        rand_var.append(randx[i])
        f_acc.append(randy[i])

rand_var_x  = np.array(rand_var_x)       
rand_var  = np.array(rand_var) 
f_acc = np.array(f_acc)
```

```python
plt.figure(2)
plt.plot(rand_var_x, f_acc, 'ro', linestyle="None", label = 'Accepted Points')
plt.plot(x,f, label = 'f(x)')
plt.xlabel('x')
plt.legend()
plt.show()
```
