```python
#ExerciseSheet3_Exercise2
import numpy as np
import matplotlib.pyplot as plt

#a)
def f(x):
    return np.sin(x)**2 - np.cos(x)**3

x = np.linspace(0, 2*np.pi, 1000)
fx = f(x)
plt.plot(x, fx, label='f(x) = sin^2(x) - cos^3(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#b)
x_random = np.random.uniform(0, 2*np.pi, 1000000)
y_random = np.random.uniform(-1, 1, 1000000)

x_accepted = []
y_accepted = []
```

```python
#c)
for i in range(1000000):
    x_value = x_random[i]
    y_value = y_random[i]
    if y_value <= f(x_value):
        x_accepted.append(x_value)
        y_accepted.append(y_value)
```

```python
#d)
plt.plot(x, fx, label='f(x) = sin^2(x) - cos^3(x)')
plt.scatter(x_accepted, y_accepted, color='red', alpha=0.5, label='Accepted Points')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#e)
plt.hist(x_accepted, bins=300, density=True, label='Histogram of Accepted x')
plt.plot(x, (fx+1)/9, label='f(x) = sin^2(x) - cos^3(x)', linewidth=2) #(fx+1)/9 to shift it to match with the normalized histogram range for visual 
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
