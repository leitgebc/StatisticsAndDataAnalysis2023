```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
c_prev = np.linspace(1e-6, 1, 1000)

p_tp_cn = 0.01
p_tp_cp = 0.6
```

```python
def p_healthy_testpos(p_tp_cn, p_tp_cp, c_prev):
    p_tn_cp = 1-p_tp_cp
    p_tn_cn = 1-p_tp_cn
    p_cp = c_prev
    p_cn_tp = (p_tp_cn*(1-p_cp))/(p_tp_cn*(1-p_cp)+p_tp_cp*p_cp)
    return p_cn_tp
```

```python
def p_healthy_testneg(p_tp_cn, p_tp_cp, c_prev):
    p_tn_cp = 1-p_tp_cp
    p_tn_cn = 1-p_tp_cn
    p_cp = c_prev
    p_cn_tn = (p_tn_cn*(1-p_cp))/(p_tn_cn*(1-p_cp) + p_tn_cp*p_cp)
    return p_cn_tn
```

```python
plt.plot(c_prev*100, p_healthy_testpos(p_tp_cn, p_tp_cp, c_prev))
plt.xlabel("Covid 19 prevalence in population [%]")
plt.ylabel("P(healthy | positive test)")
```

```python
plt.plot(c_prev*100, p_healthy_testneg(p_tp_cn, p_tp_cp, c_prev))
plt.xlabel("Covid 19 prevalence in population [%]")
plt.ylabel("P(healthy | negative test)")
```
